```python
# import pandas
import pandas as pd
```

```python
import os
```

```python
# function to save modified frames
def save_data_frame(df, filepath):
    df.to_csv(encoding='iso-8859-1', sep=';', path_or_buf=filepath, index=False)
```

```python
# not working with jupyter
# base_path = os.path.dirname(os.path.realpath(__file__))
base_path = os.path.join(os.getcwd(), 'data')


# operation parameters
operation = 'filter_version_statuses'
filename_no_type_postfix = 'ff4-to-versions-statuses-cleaned3'
file_has_header = None # use 0 for has header and None for does not have a header

raw_path = os.path.join(base_path, 'raw')
modified_path = os.path.join(base_path, 'modified')
# add subdir dropna
modified_path = os.path.join(modified_path, operation)

# to_version_statuses
to_version_statuses_headings_path = os.path.join(raw_path, 'ff4-to-versions-statuses-cleaned0.csv')
to_version_statuses_raw_path = os.path.join(raw_path, filename_no_type_postfix + '.csv')
to_version_statuses_modified_path = os.path.join(modified_path, filename_no_type_postfix + '-v2' + '-modified-' + operation + '.csv')
```

```python
# load file header


to_version_statuses_frame_headers = pd.read_csv(filepath_or_buffer=to_version_statuses_headings_path, 
                                                header=0, encoding='iso-8859-1', sep=';', nrows = 1)
```

```python
# load to data frame

to_version_statuses_frame = pd.read_csv(filepath_or_buffer=to_version_statuses_raw_path, 
                                        header=file_has_header, encoding='iso-8859-1', sep=';', names = list(to_version_statuses_frame_headers))
```

```python
# initial data inspection
########################################################## HU_VERSIONS_FRAME ###################################################
#print(to_version_statuses_frame.info(null_counts=True))
#to_version_statuses_frame
```

```python
# comments indicate which columns are retained 
to_version_statuses_frame.drop(['foreign_key',
                                'dwh_masterclient_id',
                                #'transportorder_id',
                                #'dwh_transportorder_version',
                                'inet_entity_type',
                                'inet_status_id',
                                'dwh_masterclient_id.1',
                                'inet_entity_id',
                                'created_date',
                                'updated_date',
                                'username',
                                #'status_date',
                                #'status_code',
                                'status_remarks',
                                'created_version',
                                'reason_codes',
                                'carrier_company_key',
                                'status_time_zone',
                                'status_zoned_datetime',
                                'job_instance_id',
                                'dwh_updated_datetime'], axis=1, inplace= True)
```

```python
to_version_statuses_frame.info(null_counts=True)
to_version_statuses_frame
```

```python
# count occurences of given filter index
from collections import Counter
```

```python
Counter(to_version_statuses_frame['status_code'])
```

```python
# filter for entries with status code 'LS110', 'LS120', 'LS130'
to_version_statuses_frame_lt140 = 
to_version_statuses_frame.loc[to_version_statuses_frame['status_code'].isin(['LS110', 'LS120', 'LS130'])]

# exploit behavior of sort_values(descending) 
# + drop_duplicates to only retain the most recent to version status and dwh version id
to_version_statuses_frame_lt140 = 
to_version_statuses_frame_lt140.sort_values(['transportorder_id', 'dwh_transportorder_version', 'status_code'], 
                                            ascending=False).drop_duplicates(['transportorder_id'])
```

```python
to_version_statuses_frame_140 = to_version_statuses_frame.loc[to_version_statuses_frame['status_code'] == 'LS140']
to_version_statuses_frame_140 = to_version_statuses_frame_140.sort_values(['transportorder_id', 'dwh_transportorder_version'], ascending=False).drop_duplicates(['transportorder_id'])
```

```python
to_version_statuses_frame_140
```

```python
to_version_statuses_frame_mixed = to_version_statuses_frame_lt140.append(to_version_statuses_frame_140, ignore_index=True)
```

```python
print(to_version_statuses_frame_mixed.info())
to_version_statuses_frame_mixed.head()
```

```python
groupby_count = to_version_statuses_frame_mixed.groupby('transportorder_id')['status_code'].apply(list).reset_index()
groupby_count['keep'] = groupby_count['status_code'].apply(lambda x: 1 if len(x) == 2 else 0)
```

```python
groupby_count = groupby_count.loc[groupby_count['keep'] == 1]
```

```python
to_version_statuses_frame_mixed = to_version_statuses_frame_mixed.loc[to_version_statuses_frame_mixed['transportorder_id'].isin(
            groupby_count['transportorder_id'])]
```

```python
to_version_statuses_frame_mixed
```

```python
# write result to file
save_data_frame(df=to_version_statuses_frame_mixed,filepath=to_version_statuses_modified_path)
```

```python
import glob
```

```python
# after all file fragments are processed, merge and process further
allFiles = glob.glob(modified_path + "/ff4-to-versions-statuses-cleaned?-v2*.csv")
# default data frame
frame = pd.DataFrame()

# list to contain all data frames to join
list_ = []
for file_ in allFiles:
    df = pd.read_csv(filepath_or_buffer=file_, 
                     header=0, encoding='iso-8859-1', sep=';')
    list_.append(df)
frame = pd.concat(list_, ignore_index=True)
```

```python
frame
```

```python
# write result to file
save_data_frame(df=frame,
                filepath=os.path.join(modified_path, 'merged', filename_no_type_postfix[:-1] + '-v2-merged-' + operation + '.csv'))
```
