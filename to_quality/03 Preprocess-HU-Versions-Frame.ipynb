```python
# import pandas
import pandas as pd
```

```python
import os
```

```python
# function to save modified frames
def save_data_frame(df, filepath):
    df.to_csv(encoding='iso-8859-1', sep=';', path_or_buf=filepath, index=False)
```

```python
# not working with jupyter
# base_path = os.path.dirname(os.path.realpath(__file__))
base_path = os.path.join(os.getcwd(), 'data')

# operation parameters
#operation = 'filter_missing_values'
operation = 'filter_version_statuses'
filename_no_type_postfix = 'ff4-hu-versions3'
file_has_header = None # use 0 for has header and None for does not have a header

raw_path = os.path.join(base_path, 'raw')
modified_path = os.path.join(base_path, 'modified')
# add subdir dropna
modified_path = os.path.join(modified_path, operation)

# hu_versions
hu_versions_headings_path = os.path.join(raw_path, 'ff4-hu-versions0.csv')
hu_versions_raw_path = os.path.join(raw_path, filename_no_type_postfix + '.csv')
hu_versions_modified_path = os.path.join(modified_path, filename_no_type_postfix + '-modified-' + operation + '.csv')

# filter status file path
filter_path = os.path.join(modified_path, 'merged', 'ff4-to-versions-statuses-cleaned-merged-filter_version_statuses.csv')
```

```python
# load file header


hu_versions_frame_headers = pd.read_csv(filepath_or_buffer=hu_versions_headings_path, 
                                header=0, encoding='iso-8859-1', sep=';', nrows = 1)
```

```python
hu_versions_frame = pd.read_csv(filepath_or_buffer=hu_versions_raw_path, 
                                header=file_has_header, encoding='iso-8859-1', sep=';', names = list(hu_versions_frame_headers))
```

```python
# load filter file
filter_frame = pd.read_csv(filepath_or_buffer=filter_path, header=0, encoding='iso-8859-1', sep=';')
```

```python
# initial data inspection
########################################################## HU_VERSIONS_FRAME ###################################################
print(hu_versions_frame.info())
hu_versions_frame
```

```python
# comments indicate which columns are retained 
hu_versions_frame.drop(['foreign_key',
                        'inet_package_id',
                        #'dwh_transportorder_version',
                        'dwh_masterclient_id',
                        #'position_level1',
                        #'position_level2',
                        'remarks',
                        'barcode',
                        'created_date', # assuming these technical dates have no influence
                        'updated_date',
                        'description',
                        'handlingunit_key',
                        'job_instance_id',
                        'dwh_updated_datetime',
                        'freight_class'], axis=1, inplace= True)
```

```python
# count occurences of given filter index
from collections import Counter
```

```python
Counter(hu_versions_frame.dwh_transportorder_version.values)
```

```python
# join filter file
hu_versions_frame = hu_versions_frame.merge(filter_frame, on=['transportorder_id', 'dwh_transportorder_version'], how='inner')
```

```python
print(hu_versions_frame.info())
hu_versions_frame
```

```python
# Rule: quantity = -1 indicates TOs to skip
hu_versions_frame = hu_versions_frame.loc[(hu_versions_frame['quantity'] != -1)
                                         ].reset_index()
```

```python
# drop artificially created index column
hu_versions_frame.drop(['index', 'dwh_transportorder_version', 'status_date', 'status_code'], axis=1, inplace=True)
```

```python
print(hu_versions_frame.info())
hu_versions_frame.head()
```

```python
# filtering potential invalid TOs according to missing values
# Rule 1: gross_weight_g has to be set for all positionlevel2 = null (because that is the outer level)
# Rule 2: stackability has to be set for all positionlevel2 = null (because that is the outer level)
# Rule 3: any dimension is missing of a handling unit
hu_versions_frame.loc[hu_versions_frame['transportorder_id'].isin(
hu_versions_frame.loc[(pd.isnull(hu_versions_frame['gross_weight_g']) & pd.isnull(hu_versions_frame['position_level2']))
                     | (pd.isnull(hu_versions_frame['stackability']) & pd.isnull(hu_versions_frame['position_level2']))
                     | (pd.isnull(hu_versions_frame['height_mm'])) 
                     | (pd.isnull(hu_versions_frame['width_mm'])) 
                     | (pd.isnull(hu_versions_frame['length_mm']))]['transportorder_id'].values)]
```

```python
# doing the opposite of the previous to retain all "valid" TOs
# note the leading "tilde" ~
hu_versions_frame_modified = hu_versions_frame.loc[~hu_versions_frame['transportorder_id'].isin(
        hu_versions_frame.loc[(pd.isnull(hu_versions_frame['gross_weight_g']) & pd.isnull(hu_versions_frame['position_level2']))
                             | (pd.isnull(hu_versions_frame['stackability']) & pd.isnull(hu_versions_frame['position_level2']))
                             | (pd.isnull(hu_versions_frame['height_mm'])) 
                             | (pd.isnull(hu_versions_frame['width_mm'])) 
                             | (pd.isnull(hu_versions_frame['length_mm']))]['transportorder_id'].values)
                                                    ]
```

```python
print(hu_versions_frame_modified.info())
hu_versions_frame_modified
```

```python
# write result to file
save_data_frame(df=hu_versions_frame_modified,filepath=hu_versions_modified_path)
```

```python
import glob
```

```python
# after all file fragments are processed, merge and process further
allFiles = glob.glob(modified_path + "/ff4-hu-versions*.csv")
# default data frame
frame = pd.DataFrame()

# list to contain all data frames to join
list_ = []
for file_ in allFiles:
    df = pd.read_csv(filepath_or_buffer=file_, 
                     header=0, encoding='iso-8859-1', sep=';')
    list_.append(df)
frame = pd.concat(list_, ignore_index=True)
```

```python
# check again whether some invalid data is present (missing gross weight on outer package, etc)
# apparently not

frame.loc[frame['transportorder_id'].isin(
        frame.loc[(pd.isnull(frame['gross_weight_g']) & pd.isnull(frame['position_level2']))
                             | (pd.isnull(frame['stackability']) & pd.isnull(frame['position_level2']))
                             | (pd.isnull(frame['height_mm'])) 
                             | (pd.isnull(frame['width_mm'])) 
                             | (pd.isnull(frame['length_mm']))]['transportorder_id'].values)
                                                    ]
```

```python
# doing the opposite of the previous to retain all "valid" TOs
# note the leading "tilde" ~
frame = frame.loc[~frame['transportorder_id'].isin(
        frame.loc[(pd.isnull(frame['gross_weight_g']) & pd.isnull(frame['position_level2']))
                             | (pd.isnull(frame['stackability']) & pd.isnull(frame['position_level2']))
                             | (pd.isnull(frame['height_mm'])) 
                             | (pd.isnull(frame['width_mm'])) 
                             | (pd.isnull(frame['length_mm']))]['transportorder_id'].values)
                                                    ]
```

```python
# write result to file
save_data_frame(df=frame,
                filepath=os.path.join(modified_path, 'merged', filename_no_type_postfix[:-1] + '-merged-' + operation + '.csv'))
```
