```python
# to preprocess categorical columns
def one_hot_encode(data_frame, column, nan_naming=False): 
    
    # one hot encoding of the particular series
    temp = pd.get_dummies(data=data_frame[column], dummy_na=nan_naming, prefix=column)
    
    if nan_naming:
        temp.rename(columns={temp.columns[-1]: '{} NaNs'.format(column)}, inplace=True)
        
    # drop the column that's no longer required
    data_frame.drop([column], inplace=True, axis=1)
    # join newly created columns
    data_frame = pd.concat([data_frame, temp], axis=1)
    
    # mandatory return statement
    return data_frame
```

```python
# alternate way to preprocess to make data easier to handel for hardware
def integer_encode(data_frame, column):
    
    unique_dict = {val: i for i, val in enumerate(list(set(data_frame[column].values)))}
    # print(unique_dict)
    
    # thanks to the set operation above, nan will always be at value 0!
    data_frame[column] = data_frame[column].apply(lambda x: unique_dict.get(x, 0))
    
    return data_frame
```
