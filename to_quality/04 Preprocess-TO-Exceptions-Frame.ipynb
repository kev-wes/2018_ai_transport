```python
# import pandas
import pandas as pd
```

```python
import os
```

```python
# function to save modified frames
def save_data_frame(df, filepath):
    df.to_csv(encoding='iso-8859-1', sep=';', path_or_buf=filepath, index=False)
```

```python
# not working with jupyter
# base_path = os.path.dirname(os.path.realpath(__file__))
base_path = os.path.join(os.getcwd(), 'data')

# operation parameters
operation = 'filter_version_statuses'

raw_path = os.path.join(base_path, 'raw')
modified_path = os.path.join(base_path, 'modified')
# add subdir dropna
modified_path = os.path.join(modified_path, operation)

# to_exceptions
to_exceptions_raw_path = os.path.join(raw_path, 'ff4-to-exceptions0.csv')
to_exceptions_modified_path = os.path.join(modified_path, 'ff4-to-exceptions0-modified-' + operation + '.csv')
```

```python
# load to data frame
to_exceptions_frame = pd.read_csv(filepath_or_buffer=to_exceptions_raw_path, 
                                  header=0, encoding='iso-8859-1', sep=';')

```

```python
# initial data inspection

print(to_exceptions_frame.info())
to_exceptions_frame
```

```python
# drop technical and unneccessary cols
to_exceptions_frame.drop(['dwh_masterclient_id',
                          'load_id',
                          'issue_type_name'], axis=1, inplace=True)
```

```python
# drop first row, is a duplicate of column headers
to_exceptions_frame.drop([0], axis=0, inplace=True)
```

```python
# reset the index
to_exceptions_frame.reset_index(inplace=True)
to_exceptions_frame.drop(['index'], axis=1, inplace=True)
```

```python
print(to_exceptions_frame.info())
to_exceptions_frame
```

```python
from collections import Counter
```

```python
Counter(to_exceptions_frame['issue_type_code'])
```

```python
# write result to file
save_data_frame(df=to_exceptions_frame,filepath=to_exceptions_modified_path)
```

```python
import glob
```

```python
# after all file fragments are processed, merge and process further
allFiles = glob.glob(modified_path + "/ff4-to-exceptions*.csv")
# default data frame
frame = pd.DataFrame()

# list to contain all data frames to join
list_ = []
for file_ in allFiles:
    df = pd.read_csv(filepath_or_buffer=file_, 
                     header=0, encoding='iso-8859-1', sep=';')
    list_.append(df)
frame = pd.concat(list_, ignore_index=True)
```

```python
to_exceptions_frame = frame
```

```python
to_exceptions_frame.info()
```

```python
# write result to file
save_data_frame(df=frame,
                filepath=os.path.join(modified_path, 'merged', 'ff4-to-exceptions' + '-merged-' + operation + '.csv'))
```
