```python
# after all file fragments are processed, merge and process further
allFiles = glob.glob(modified_path + "/ff4-to-versions-c0leaned*.csv")
# default data frame
frame = pd.DataFrame()

# list to contain all data frames to join
list_ = []
for file_ in allFiles:
    df = pd.read_csv(filepath_or_buffer=file_, 
                     header=0, encoding='iso-8859-1', sep=';')
    list_.append(df)
frame = pd.concat(list_, ignore_index=True)
```

```python
# write result to file
save_data_frame(df=frame,
                filepath=os.path.join(modified_path, 'merged', filename_no_type_postfix[:-1] + '-merged-' + operation + '.csv'))
```

```python
# drop artificially created index column
to_versions_frame.drop([#'transportorder_id',
                        'dwh_transportorder_version'
                        #'created_date',
                        #'updated_date',
                        #'incoterm',
                        #'owner_company_key',
                        #'consignor_location_key',
                        #'recipient_location_key',
                        #'pickup_date_from',
                        #'pickup_date_to',
                        #'delivery_date_from',
                        #'delivery_date_to',
                        #'steering_partner_id',
                        #'pickup_location_key',
                        #'delivery_location_key',
                        #'status_date',
                        #'status_code'
                        ], axis=1, inplace=True)
```

```python
print(to_versions_frame.info(null_counts=True))
to_versions_frame.head()
```

```python
# filtering potential invalid TOs according to missing values
# CURRENTLY IGNORING RULE 1
# Rule 1: pickup and delivery dates have to be set
# Rule 2: pickup and deliver location keys have to be set
to_versions_frame.loc[to_versions_frame['transportorder_id'].isin(
    to_versions_frame.loc[#pd.isnull(to_versions_frame['pickup_date_from']) 
                         #| pd.isnull(to_versions_frame['pickup_date_to'])
                         #| pd.isnull(to_versions_frame['delivery_date_from'])
                         #| pd.isnull(to_versions_frame['delivery_date_to'])
                         #|
                         pd.isnull(to_versions_frame['pickup_location_key'])
                         | pd.isnull(to_versions_frame['delivery_location_key'])]['transportorder_id'].values)]
```

```python
# doing the opposite of the previous to retain all "valid" TOs
# note the leading "tilde" ~
to_versions_frame_modified = to_versions_frame.loc[~to_versions_frame['transportorder_id'].isin(
    to_versions_frame.loc[#pd.isnull(to_versions_frame['pickup_date_from']) 
                         #| pd.isnull(to_versions_frame['pickup_date_to'])
                         #| pd.isnull(to_versions_frame['delivery_date_from'])
                         #| pd.isnull(to_versions_frame['delivery_date_to'])
                         #|
                          pd.isnull(to_versions_frame['pickup_location_key'])
                         | pd.isnull(to_versions_frame['delivery_location_key'])]['transportorder_id'].values)
                                                  ]
```
